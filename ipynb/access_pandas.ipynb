```python
import pyodbc
import pandas as pd
import numpy as np

conn = pyodbc.connect(r'Driver={Microsoft Access Driver (*.mdb, *.accdb)};DBQ=C:\\Users\\suyu\\Desktop\\AI\\company.accdb')

SQL = '''\
SELECT 
EMPLOYEE.[Fname] AS 名, EMPLOYEE.[Lname] AS 姓,
DatePart('yyyy',EMPLOYEE.[Bdate]) AS 出生年份,
DateDiff('yyyy',EMPLOYEE.[Bdate],Date()) AS 年齡1,
DateDiff('d',EMPLOYEE.[Bdate],Date()) AS 年齡2
FROM EMPLOYEE
'''

cur = conn.cursor()
cur.execute(SQL)
list = cur.fetchall()

desc = cur.description
cur.close()
conn.close()

df_column = pd.DataFrame(np.array(desc))
df = pd.DataFrame(np.array(list), columns=df_column[0])
df
```
