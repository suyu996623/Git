```python
import sqlite3
import pandas as pd

database = sqlite3.connect('/content/drive/MyDrive/Colab Notebooks/03 用excel學python(書本範例)/CH09.sqlite') # 建立資料庫連線
cursor = database.cursor()
表格 =   cursor\
        .execute("select name from sqlite_master where type = 'table' order by name")\
        .fetchall()

DF串列 = []
for item in 表格 :
    資料 = cursor.execute(f"select * from {item[0]}").fetchall()
    欄位 = [tuple1[0] for tuple1 in cursor.description]
    DF串列.append( pd.DataFrame(資料,columns=欄位) )

database.close()  # 關閉資料庫連線

學生 = DF串列[0]
教師 = DF串列[1]
系所 = DF串列[2]
課程 = DF串列[3]
選課單 = DF串列[4]
```

```python
# 1. 列出所有選課資料的  選課序號，學年，學生姓名，學生所屬學系，科目名稱，教師姓名，上課教室

table = 選課單.merge(學生, left_on='選修學生', right_on='學號')
table = table.merge(課程, left_on='選修課程', right_on='科目代碼')
table = table.merge(教師, left_on='任課老師', right_on='教師编號')
table = table.merge(系所, left_on='系所', right_on='代碼')
table = table[['選課序號','學年','姓名_x','學系','科目名稱','姓名_y','上課教室']]
table.rename(columns={'姓名_x':'學生姓名','學系':'學生所屬學系','姓名_y':'教師姓名'}).sort_values('選課序號')
```

```python
# 2. 列出所有學生的學號, 姓名, 生日, 年齡 (動態新增年齡欄位)

import datetime as dt

def func1(x) :

    birth = x.split()[0]
    x1 = dt.datetime.strptime(birth,"%Y-%m-%d")
    x2 = dt.datetime.today()
    if (x1.month < x2.month) or (x1.month == x2.month and x1.day <= x2.month) :
        #生日已過
        age = x2.year - x1.year
    else :
        age = x2.year - x1.year - 1
    return age

學生['年齡'] = 學生['生日'].apply(func1)
學生[['學號','姓名','生日','年齡']]
```
