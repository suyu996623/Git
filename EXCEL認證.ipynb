```python
# numpy.random.randn -- Return a sample (or samples) from the “standard normal” distribution.

weather_df = pd.DataFrame(np.random.rand(10,2)*5,
                          index=pd.date_range(start="2021-01-01", periods=10),
                          columns=["Tokyo", "Beijing"])

def rain_condition(v):
    if v < 1.75:
        return "Dry"
    elif v < 2.75:
        return "Rain"
    return "Heavy Rain"

def make_pretty(styler):
    styler.set_caption("Weather Conditions")  #  Set the text added to a <caption> HTML element.
    styler.format(formatter=rain_condition)  #  Format the text display value of cells.
    styler.format_index(formatter=lambda v: v.strftime("%A"))  #  Format the text display value of index labels or column headers.
    styler.background_gradient(axis=None, vmin=1, vmax=5, cmap="Blues")  #  Color the background in a gradient style.
    #  to the entire DataFrame at once with axis=None.
    #  Minimum data value that corresponds to colormap minimum value.
    #  Maximum data value that corresponds to colormap maximum value.
    #  Matplotlib colormap.
    return styler

weather_df
```

```python
#  property DataFrame.style    Returns a Styler object.
#  Styler properties   Style application  https://pandas.pydata.org/docs/reference/style.html
#  Styler.pipe(func, *args, **kwargs)   Apply func(self, *args, **kwargs), and return the result.

weather_df.loc["2021-01-04":"2021-01-08"].style.pipe(make_pretty)
```

```python
# https://towardsdatascience.com/a-quick-and-easy-guide-to-conditional-formatting-in-pandas-8783035071ee

def style_2(one_row, props=''):
    # print(s['2014年業績'],s['2015年業績'])
    # print(props)
    return ['color:red' for item in one_row] if one_row['2014年業績'] < one_row['2015年業績'] else ['color:green' for item in one_row]

df1.style.apply(style_2, axis=1).format(precision=2, thousands=",", decimal=".")  # 認證第六題
```
